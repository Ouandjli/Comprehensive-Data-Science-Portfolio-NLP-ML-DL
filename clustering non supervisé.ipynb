```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from scipy.optimize import linear_sum_assignment
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster




#
# Unsupervised Classification: the K-Mean Algorithm
# THe classic Iris data set
#

#DisplayPlots = True

# Iris data set
#X: Sepal length, Sepal width, Petal length, Petal width
#Y: 0 = setosa, 1 = versicolor, 2 = virginica

iris = load_iris()
X = iris.data
print("X=",X)
Y = iris.target
print("Y=",Y)
print("length=",len(X))

# Créer la matrice de confusion entre les labels réels et les labels K-Means
kmeans = KMeans(n_clusters=3, random_state=100)
kmeans.fit(X)
labels = kmeans.labels_
conf_matrix = confusion_matrix(Y, labels)
print("Confusion Matrix:\n", conf_matrix)

# Utiliser l'algorithme de l'assignation hongroise pour trouver la correspondance optimale
row_ind, col_ind = linear_sum_assignment(-conf_matrix)

# Calculer le nombre de hits
hits = conf_matrix[row_ind, col_ind].sum()
total_samples = len(Y)
hit_ratio = hits / total_samples

print(f"Hit Ratio: {hit_ratio:.2f}")

kmeans = KMeans(n_clusters=3, random_state=100)
kmeans.fit(X)
labels = kmeans.labels_

print("labels=", labels)
conf_matrix = confusion_matrix(Y, labels)
print("Confusion Matrix:\n", conf_matrix)
#Utiliser l'algorithme de l'assignation hongroise pour trouver la correspondance optimale
row_ind, col_ind = linear_sum_assignment(-conf_matrix)

#calculer le hit-ratio
hits = conf_matrix[row_ind, col_ind].sum()
total_samples = len(Y)
hit_ratio = hits / total_samples

print(f"Hit Ratio: {hit_ratio:.2f}")

# apprentissage non-supervisé, par k-moyennes (k-means algorithm)

kmeans = KMeans(n_clusters=3, random_state=100)
kmeans.fit(X)
labels = kmeans.labels_
print("labels=",labels)

# Appliquer l'algorithme K-Means pour différentes valeurs de k
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=100)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Appliquer K-Means avec le nombre optimal de clusters (k=3 dans cet exemple)
optimal_k = 3
kmeans_optimal = KMeans(n_clusters=optimal_k, random_state=100)
kmeans_optimal.fit(X)
labels_optimal = kmeans_optimal.labels_
print("Optimal labels=", labels_optimal)



# 2D projections
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
centers_reduced = pca.transform(kmeans.cluster_centers_)

# 3D projection
pca = PCA(n_components=3)
X_reduced_3d = pca.fit_transform(X)
centers_reduced_3d = pca.transform(kmeans.cluster_centers_)


DisplayPlots = True  # Définir DisplayPlots sur True pour activer les visualisations

if DisplayPlots:
    # 2D plot
    plt.figure()
    plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, cmap='viridis', marker='o')
    plt.scatter(centers_reduced[:, 0], centers_reduced[:, 1], s=300, c='red', marker='X')
    plt.xlabel('P.C. 1')
    plt.ylabel('P.C. 2')
    plt.title('K-means Clustering -- Iris Dataset (2D PCA)')

    plt.show()

    # 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    scatter = ax.scatter(X_reduced_3d[:, 0], X_reduced_3d[:, 1], X_reduced_3d[:, 2], c=labels, cmap='viridis')
    ax.scatter(centers_reduced_3d[:, 0], centers_reduced_3d[:, 1], centers_reduced_3d[:, 2], s=300, c='red', marker='X')
    ax.set_xlabel('P.C. 1')
    ax.set_ylabel('P.C. 2')
    ax.set_zlabel('P.C. 3')
    ax.set_title('K-means Clustering -- Iris Dataset (3D PCA)')
    legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")
    ax.add_artist(legend1)

    plt.show()

# hierarchical clustering
# Charger le dataset Iris
# Charger le dataset Iris
iris = load_iris()
X = iris.data
Y = iris.target

print("X=", X)
print("Y=", Y)
print("length=", len(X))

# Méthodes de clustering
methods = ['single', 'complete', 'average']

# Fonction pour tracer le dendrogramme et afficher les clusters
def plot_dendrogram_and_clusters(X, method, num_clusters=3):
    Z = linkage(X, method=method)

    plt.figure(figsize=(10, 5))
    dendrogram(Z, labels=Y, leaf_rotation=90, leaf_font_size=10, color_threshold=0)
    plt.title(f'Dendrogram for {method} linkage')
    plt.xlabel('Sample Index or (Cluster Size)')
    plt.ylabel('Distance')
    plt.show()

    # Déterminer les clusters
    clusters = fcluster(Z, t=num_clusters, criterion='maxclust')

    # Visualiser les clusters
    pca = PCA(2)
    X_pca = pca.fit_transform(X)

    plt.figure(figsize=(10, 5))
    for i in range(1, num_clusters + 1):
        plt.scatter(X_pca[clusters == i, 0], X_pca[clusters == i, 1], label=f'Cluster {i}')

    plt.title(f'Clusters formed by {method} linkage')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend()
    plt.show()

# Tracer les dendrogrammes et afficher les clusters pour chaque méthode
for method in methods:
    plot_dendrogram_and_clusters(X, method)


```
